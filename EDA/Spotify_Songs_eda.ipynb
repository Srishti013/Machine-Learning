# Data Analysis

## Top 5 Popular Artists

```python
Top_5_artists = df.groupby("artist").count().sort_values(by='song_title',ascending=False)['song_title'][:5]
Top_5_artists
```

```python
Top_5_artists.plot.barh()
plt.show()
```

## Top 5 Loudest Tracks

```python
Top_5_loudest = df[['loudness','song_title']].sort_values(by='loudness')[:5]
Top_5_loudest
```

```python
sns.barplot(x='loudness',y='song_title',data = Top_5_loudest)
plt.title("Top 5 loudest tracks")
plt.show()
```

## Artist with most danceability song

```python
Most_danceability_artist = df[['artist','danceability']].sort_values(by='danceability', ascending=False)[:1]
Most_danceability_artist['artist']
```

## Top 10 instrumental tracks

```python
Top_10_instrumental_tracks = df[['song_title','instrumentalness']].sort_values(by='instrumentalness',ascending=False)[:10]
Top_10_instrumental_tracks
```

```python
plt.pie(x="instrumentalness",data=Top_10_instrumental_tracks,autopct='%1.2f%%', labels=Top_10_instrumental_tracks.song_title )
plt.title('Top 10 instrumental tracks')
plt.show()
```

## Feature Plots

```python
cols = ['tempo','loudness','acousticness','danceability','duration_ms','energy','instrumentalness','liveness','speechiness','valence']
```

```python
for col in cols:
  pos_data = df[df['target']==1][col]
  neg_data = df[df['target']==0][col]

  plt.figure(figsize=(12,7))

  sns.distplot(pos_data, bins=30, label='Positive', color='green')
  sns.distplot(neg_data, bins=30, label='Negative', color='red')

  plt.legend(loc='upper right')
  plt.title(f"Positive And Negative Histogram Plot For {col}")
  plt.show()
```
